```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

### Random Walk

```python
def random_walk(N):
    start = 0
    walk = [start]
    for n in range(N):
        step = np.random.choice([-1,1], p=[0.5,0.5])
        start = start + step
        walk.append(start)
        
    return np.array(walk)
```

```python
walk = random_walk(10_000)
```

```python
plt.figure(figsize=(10,5))
plt.plot(walk)
plt.xlabel('Time',fontsize=20)
plt.ylabel(r'$State$',fontsize=20)
```

### Independet Process

```python
states = np.arange(1, 5, 1)
states
```

```python
probs = np.array([0.5, 0.25, 0.12, 0.13])
probs.sum()
```

```python
def independent_process(states, probabilities, N):
    samples = []
    for n in range(N):
        state = np.random.choice(states, p=probabilities)
        samples.append(state)
        
    return np.array(samples)
```

```python
process = independent_process(states, probs, 50)

plt.plot(process)
plt.xlabel('Time',fontsize=20)
plt.ylabel(r'$State$',fontsize=20)
```

### Markov process

```python
states = np.arange(1, 5, 1)
states
```

```python
s1=[0,1.0,0,0]
s2=[0.5,0,0.25,0.25]
s3=[0,0.5,0,0.5]
s4=[0,0.5,0.5,0]
transitions = np.array([s1,s2,s3,s4])
transitions
```

```python
def graph_process(states, transitions, N, initial):
    state = initial
    samples = []
    for n in range(N):
        #print(transitions[state-1])
        state = np.random.choice(states, p=transitions[state-1])
        samples.append(state)
        
    return np.array(samples)
```

```python
process = graph_process(states, transitions, 30, 1)

plt.plot(process)
plt.xlabel('Time',fontsize=20)
plt.ylabel(r'$State$',fontsize=20)
```
